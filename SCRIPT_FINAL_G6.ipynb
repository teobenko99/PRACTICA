##4.5 Errores tipográficos en variables categóricas

Se verifica si en los sub-niveles de las variables categóricas, existen valores registrados de forma diferente que representan lo mismo y en tal caso, se procede a unificarlos.

Se utilizarán gráficos de barras para el análisis.

```python
# Selección de variables categóricas
col_cat = data.select_dtypes(include="object").columns.tolist()

# Crear gráficos de barras
fig, ax = plt.subplots(nrows=len(col_cat), ncols=1, figsize=(10, 5 * len(col_cat)))
fig.subplots_adjust(hspace=0.5)

for i, col in enumerate(col_cat):
    sns.countplot(x=col, data=data, ax=ax[i], order=data[col].value_counts().index)
    ax[i].set_title(f"Distribución de: {col}")
    ax[i].set_xticklabels(ax[i].get_xticklabels(), rotation=30)

plt.tight_layout()
plt.show()
```

**No** se observan sub-niveles en ninguna de las variables categóricas que esten escritos de diferentes maneras y representen lo mismo.

Se observa que la mayoría de los sub-niveles estan escritos con mayúscula inicial. Otros se escriben todo minúscula y otras variables tienen escrito un sub-nivel con mayúscula inicial y otro sub-nivel todo en minúscula. **Dada esta osbervación, y solamente con el fin estético de unificar la forma de escribir, se pasarán todos los sub-niveles a minúscula.**

```python
# Escribimos todos los sub-niveles en minúscula usando str.lower
for column in data.columns:
    if column in col_cat:
        data[column] = data[column].str.lower()

# Se generan las gráficas nuevamente
fig, ax = plt.subplots(nrows=len(col_cat), ncols=1, figsize=(10, 5 * len(col_cat)))
fig.subplots_adjust(hspace=0.5)

for i, col in enumerate(col_cat):
    sns.countplot(x=col, data=data, ax=ax[i], order=data[col].value_counts().index)
    ax[i].set_title(f"Distribución de: {col}")
    ax[i].set_xticklabels(ax[i].get_xticklabels(), rotation=30)

plt.tight_layout()
plt.show()
```

## 6.2 Análisis univariado: relación de cada variable predictora con la variable a predecir

```python
# Definimos variables categóricas a analizar
cat_vars = ['Gender', 'Customer Type', 'Type of Travel', 'Class']

# Graficamos cada situación
for col in cat_vars:
    plt.figure(figsize=(8, 4))
    sns.countplot(x=col, hue='satisfaction', data=data, order=data[col].value_counts().index)
    plt.title(f'Satisfacción según {col}')
    plt.xticks(rotation=30)
    plt.tight_layout()
    plt.show()
```

```python
#Observamos la distribución mediante tablas
cat_vars = ['Gender', 'Customer Type', 'Type of Travel', 'Class']

for var in cat_vars:
    print(f"\n*Distribución de satisfacción según {var}:\n")
    tabla = pd.crosstab(data[var], data['satisfaction'], normalize='index') * 100
    print(tabla.round(2))
```

* **Clase:**

  * Nota: Si el segmento 'eco' y 'eco plus' son los más utilizados, se deberían tomar acciones con urgencia, dado el nivel de insatisfacción. No se observa gran diferencia entre estas dos categorías como para justificar pagar un precio más elevado por una que por la otra.

Esto muestra que hay diferencias en la distribución de la satisfacción según distintos rangos etarios. Parecería que los jóvenes y adultos mayores, suelen estar mas insatisfechos que los adultos. Por lo tanto, segmentamos la variable.

```python
# Definimos el criterio para la segmentación
def clasificar_edad(a):
    if a <= 35:
        return 'Jovenes'
    elif a <= 65:
        return 'Adultos'
    else:
        return 'Adultos mayores'

# Aplicamos la segmentación para definir la nueva variable "Tipo de Vuelo"
data['rango_edad'] = data['Age'].apply(clasificar_edad)

#Graficamos
sns.countplot(data=data, x='rango_edad', hue='satisfaction')
plt.title('Distribución de satisfacción por grupo de edad')
plt.xlabel('Grupo de Edad')
plt.ylabel('Cantidad')
plt.legend(title='Satisfacción')
plt.show()

pd.crosstab(data['rango_edad'], data['satisfaction'], normalize='index') * 100
```

Criterios elegidos:

* Corto: hasta 1500 millas
* Medio: de 1 500 a 4 000 millas
* Largo: más de 4 000 millas

Variable a definir: 'Tipo de Vuelo' (categórica)

```python
# Definimos el criterio para la segmentación
def clasificar_distancia(dist):
    if dist <= 1500:
        return 'Corto'
    elif dist <= 3000:
        return 'Medio'
    else:
        return 'Largo'

# Aplicamos la segmentación para definir la nueva variable "Tipo de Vuelo"
data['Tipo de vuelo'] = data['Flight Distance'].apply(clasificar_distancia)

#Graficamos
sns.countplot(data=data, x='Tipo de vuelo', hue='satisfaction')
plt.title('Distribución de satisfacción por el Tipo de vuelo')
plt.xlabel('Tipo de vuelo')
plt.ylabel('Cantidad')
plt.legend(title='Satisfacción')
plt.show()

pd.crosstab(data['Tipo de vuelo'], data['satisfaction'], normalize='index') * 100
```

* Los viajes cortos, tienen mayor cantidad de pasajeros insatisfechos (64%).
* Los viajes con distancia media y larga, tienen mayor cantidad de pasajeros satisfechos (64% y 78% respectivamente)

* Nota: Esto podría indicar que hay determinadas variables que se perciben mejor cuando el vuelo es más lago (por ejemplo, mejor comida, mayor entretenimiento, mejor servicio general, etc). Por lo tanto, puede considerarse como variable predictora pero no tiene un impacto tan fuerte como las otras que se estuvieron evaluando.

```python
# grafiquemos la satisfacción en función de la demora en la salida y en la llegada

sns.histplot(data=data, x='Departure Delay in Minutes', hue='satisfaction_bin', kde=True, element='step', stat='density')
plt.title("Distribución de demoras en la salida por satisfacción")
plt.xlabel("Demoras en la salida")
plt.ylabel("Densidad")
plt.show()

sns.histplot(data=data, x='Arrival Delay in Minutes', hue='satisfaction_bin', kde=True, element='step', stat='density')
plt.title("Distribución de demoras en la llegada por satisfacción")
plt.xlabel("Demoras en la llegada")
plt.ylabel("Densidad")
plt.show()
```

Las demoras extremas, es decir, los outliers que observamos en las secciones previas, estan ocultando las diferencias existentes. Por lo tanto, vamos a segmentar ambas variables.

Criterios elegidos:
* Sin Demora: 0 minutos

* Leve: de 1 a 15 min

* Moderada: de 16 a 60 minutos

* Significativa de 61 a 180 minutos

* Severa: más de 181 minutos

Variables a definir: 'Delay salida' (categórica) y 'Delay llegada' (categórica)

```python
# Definimos el criterio para la segmentación de Departure Delay y Arrival Delay
def clasificar_delay(mins):
    if mins == 0:
        return 'Sin demora'
    elif mins <= 15:
        return 'Leve (1–15 min)'
    elif mins <= 60:
        return 'Moderada (16–60)'
    elif mins <= 180:
        return 'Significativa (61–180)'
    else:
        return 'Severa (>180)'

# Aplicamos la segmentación para definir las nuevas variables "Delay salida" y "Delay llegada"
data['Delay salida'] = data['Departure Delay in Minutes'].apply(clasificar_delay)
data['Delay llegada'] = data['Arrival Delay in Minutes'].apply(clasificar_delay)

#Graficamos
plt.figure(figsize=(10, 5))
orden = ['Sin demora', 'Leve (1–15 min)', 'Moderada (16–60)', 'Significativa (61–180)', 'Severa (>180)']
sns.countplot(data=data, x='Delay salida', hue='satisfaction', order=orden)
plt.title('Distribución de satisfacción por el restraso a la salida')
plt.xlabel('Delay salida')
plt.ylabel('Cantidad')
plt.legend(title='Satisfacción')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()

print(pd.crosstab(data['Delay salida'], data['satisfaction'], normalize='index') * 100)

plt.figure(figsize=(10, 5))
sns.countplot(data=data, x='Delay llegada', hue='satisfaction', order=orden)
plt.title('Distribución de satisfacción por el restraso a la llegada')
plt.xlabel('Delay llegada')
plt.ylabel('Cantidad')
plt.legend(title='Satisfacción')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()

print(pd.crosstab(data['Delay llegada'], data['satisfaction'], normalize='index') * 100)
```

De esta manera, logramos visualizar mejor la distribución general sin que los outliers extremos opaquen los datos comunes. **La tendencia indica que no suele haber demoras en la salida y en la entrada, y de existir, suelen ser leves o moderadas, con un mayor impacto en la salida.**

  * Las demoras se presentan de manera bastante **lineal**, *a mayor retraso en la llegada o en la salida, mayor cantidad de pasajeros insatisfechos*. Esto estaría indicando, que **las demoras son factores críticos** que afectan a la satisfacción de los pasajeros, siendo mayor en la llegada que en la salida.

  * Cuando **no** hay demora, los valores son prácticamente iguales, lo que podría indicar que **cuando no hay demora, la satisfacción no parecería verse afectada por esta variables.**

Un punto importante a remarcar es que este conjunto de datos análizado, no tiene información sobre qué genera la demora, si fue una cuestión del aeropuerto, un pasajero, una falla operativa de la compañia aérea, mantenimiento del avión, etc. Ademas, la tendencia indica que no suele haber demoras, y que en estos casos, no parece existir relación con la satisfacción. Por lo tanto, concluimos no considerar en este análisis a las demoras como una variable predictora.

## 6.3 Análisis bivariado: relación de pares de variables predictoras con la variable a predecir

Dado que las combinaciones posibles a realizar son infinitas, vamos a centrarnos en las conclusiones principales arribadas en el análisis univariado sobre las variables claves que resultarían ser predictoras, para realizar un análisis bivariado completo que nos permita encontrar relaciones.

Primero definimos la función que vamos a utilizar para hacer los distintos gráficos bivariados.

```python
def graficar_satisfaccion_bivariada(col_list, orden=None):
    tc = data.groupby(col_list)['satisfaction_bin'].mean().to_frame().reset_index()

    # Gráfica
    plt.figure(figsize=(7, 5))
    ax = sns.pointplot(
        x='satisfaction_bin', y=col_list[0], hue=col_list[1],
        data=tc, join=False, order=orden
    )
    ax.yaxis.grid(True)
    ax.xaxis.grid(True)
    plt.title(f'Satisfacción promedio para {col_list[0]} y {col_list[1]}')
    plt.xlabel('Porcentaje de satisfacción')
    plt.xlim((0, 1))
    plt.show()
```

En segundo lugar, vamos a realizar una transformación de las variables que representan a los servicios puntuados por los pasajeros para una mejor interpretación de los resultados, dado que la escala que se utiliza es de 0 a 5 y la dispersión de los resultados puede dificultar la comprensión de lo obtenido en los diferentes gráficos.

```python
# Definimos la segmentación:
def clasificar_servicios(a):
    if a <= 2:
        return 'Bajo'
    elif a <= 4:
        return 'Medio'
    else:
        return 'Alto'
```
